```python
import pandas as pd
import numpy as np
from tabula import read_pdf
```

```python
url = r"C:\Users\gomes\Downloads\Word frequency in English (Corpus of Contemporary American English) (Z-Library).pdf"
```

```python
df = read_pdf(
    url, 
    pages='1-1112', 
    pandas_options={
        "header": None, 
        "names":['Classe', 'Palavra', 'Total', 'Spoken'],
        }, 
    multiple_tables=False, 
    )[0]
```

```python
df.Classe = df.Classe.str[-1]
```

```python
df['Letra'] = df.Palavra.str[0]
```

```python
df['Letra2'] = df.Palavra.str[:2]
```

```python
df
```

```python
df.Letra.value_counts(ascending=True).plot(kind='barh')
```

```python
df.Letra2.value_counts(ascending=True).plot(kind='barh', figsize=(10,50))
```
